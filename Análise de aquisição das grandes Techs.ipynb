# -Analisando dados de aquisições de BigTechs.

Farei uma breve análise das aquisições das maiores companhias Tech usando um dataset que encontei no Kaggle.

```python
#Importando as Bibliotecas que serão úteis.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Biblioteca para ignorar avisos.
import warnings
warnings.filterwarnings('ignore')
```

```python
#Fazendo a leitura dos dados.
bd = pd.read_csv('Biggest Tech Acquisitions 2020.csv')

#Verificando a leitura
bd
```

```python
#Verificando a dimensão do dataset

bd.shape
```

```python
#Verificando se há campos nulos.

bd.isnull()
```

```python
#Verificando compos únicos

bd.nunique()
```

```python
#Estatísticas descritivas básicas.

bd.describe().round()
```

```python
#Criando um BoxPlot para melhor visualização.

plt.figure(figsize=(16,7))
plt.title('Preço de Custo da Empresas Aquisicionadas', fontsize = 14)
sns.barplot( data=bd , x='Acquirer', y='Amount in billions (USD)')


#Editando a linha e a coluna do gráfico.

plt.xticks(rotation = 45, ha = 'right')
plt.ylabel('Bilhões de dólares', fontsize = 12)
plt.xlabel('Compradoras', fontsize = 12);
```

```python
#Grafíco de densidade do valor das aquisições.

sns.kdeplot(bd['Amount in billions (USD)'], shade = True, bw = 1 ,color ='#708090', linewidth = 2.5);
```
